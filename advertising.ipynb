```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv("C:\\Users\\Zahidul Islam\\Downloads\\Advertising.csv")
```

```python
df.head()
```

```python
df['Total Spent']=df['TV']+df['newspaper']+df['radio']
```

```python
sns.scatterplot(data=df,x='Total Spent',y='sales')
```

```python
sns.regplot(data=df,x='Total Spent',y='sales')
```

```python
x= df['Total Spent']
y= df['sales']
```

```python
x1= sm.add_constant(x)
result= sm.OLS(y,x1).fit()
```
